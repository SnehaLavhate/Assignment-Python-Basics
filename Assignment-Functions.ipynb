1. What is the difference between a function and a method in Python?

     #### Function :
   * A function is a block of reusable code that performs a specific task.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

    * map() - Execuetes a specified function for each of iter of an iterable.
    * reduce() - It does the folding of the sequence , or applies to  a function to all elements in a sequence ,reducing it to a single value.
    * filter() - used to filter out items from an iterable based on the condition. It keeps only the elements which the function returns True.
      

```python
#map():
def add(x):
    return x+1
l = [1,2,3,4,5,6]
map(add,l)
list(map(add,l))
```

```python
#reduce():
from functools import reduce  # reduce is not a Python keyword , so we have to import it from a library.
l = [1,2,3,4,5,6,7]
reduce(lambda x,y : x+1,l)
```

```python
#filter():
l = [2,3,5,6,7,8]
list(filter(lambda x : x % 2 == 0, l))
```

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_eve(*numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total = total + num
    return f"The sum of all even numbers in the list is :{total}"
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
import math

def is_prime(n):
  """Checks if a number is prime."""
  if n <= 1:
    return False
  for i in range(2, int(math.sqrt(n)) + 1):
    if n % i == 0:
      return False
  return True

# Check numbers from 1 to 200
for number in range(1, 201):
  if is_prime(number):
    print(f"{number} is a prime number.")
  else:
    print(f"{number} is not a prime number.")
    
```

```python
#5. Create an iterator class in Python that generates the Fibonaccci sequence up to specified number of terms.
class FibonacciIterator:
    def __init__(self,num_terms):
        self.num_terms = num_terms
        self.count = 0
        self.a,self.b = 0,1
    def __iter__(self):
        return self
    def __next__(self):
        if self.count>=self.num_terms:
            raise StopIteration
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a,self.b = self.b,self.a+self.b
            self.count += 1
            return self.b
```

```python
fib = FibonacciIterator(10)
for i in fib:
    print(i)
```

```python
#6. Write a generator function in Python that yields the powers of 2 upto a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent
```

```python
for power in powers_of_two(6):
    print(power)
```
